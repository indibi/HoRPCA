```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,2*np.pi, 200)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid
plt.show()
```

```python
fig = plt.figure()  # an empty figure with no Axes
fig, ax = plt.subplots()        # a figure with single axes
fig, axs = plt.subplots(2,2)    # a figuer with 2x2 grid of axes
plt.show()
```

```python
np.random.seed(16593650)
data = {'a': np.arange(50),
        'c': np.random.randint(0,50,50),
        'd': np.random.randn(50)}
data['b'] = data['a']+10*np.random.randn(50)
data['d'] = np.abs(data['d'])*100

fig,ax = plt.subplots(figsize=(5,2.7), layout='constrained')
ax.scatter('a', 'b', c='c', s='d', data=data)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')

```

```python
# Object oriented example.
x = np.linspace(0,2,100)

# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.
fig, ax = plt.subplots(figsize=(5,2.7), layout='constrained')
ax.plot(x,x,label='Linear')
ax.plot(x,x**2,label='Quadratic')
ax.plot(x,x**3,label='Cubic')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Simple Plot')
ax.legend();
```

```python
x = np.linspace(0,2,100)

plt.figure(figsize=(5,2.7), layout='constrained')
plt.plot(x,x,label='linear')
plt.plot(x,x**2, label='Quadratic')
plt.plot(x,x**3, label='Cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

```python
# Making Elper Functions ELP

def my_plotter(ax, data1, data2, param_dict):
    """_summary_

    Args:
        ax (_type_): _description_
        data1 (_type_): _description_
        data2 (_type_): _description_
        param_dict (_type_): _description_
    """
    out = ax.plot(data1, data2, **param_dict)
    return out


data1, data2, data3, data4 = np.random.randn(4,100) # Make 4 random data sets
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(5,2.7))
my_plotter(ax1, data1, data2, {'marker': 'x', 'color':'red'})
my_plotter(ax2, data2, data4, {'marker': 'o', 'color':'violet'})
```

```python
# Styling Artists
fig, ax= plt.subplots(figsize=(6,3))
x = np.arange(len(data1))
ax.plot(x,np.cumsum(data1), color='blue', linewidth=3, linestyle='--')
l, = ax.plot(x, np.cumsum(data2), color='red', linewidth=2)
l.set_linestyle('-')
```

```python
# Colors
fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(data1,data2,s=50, facecolor='C0', edgecolor='k');
```

```python
# Linewidths, linestyles, and markersizes
fig, ax = plt.subplots(figsize=(7,4))
ax.plot(data1, 'o', label='data1')
ax.plot(data2, 'd', label='data2')
ax.plot(data3, 'v', label='data3')
ax.plot(data4, 's', label='data4')
ax.legend()
```
